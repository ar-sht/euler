```python
maximum = 0
print("HORIZONTAL")
# horizontal line loop
for row in nums:
    for i in range(len(row) - 3):
        ns = [row[i], row[i+1], row[i+2], row[i+3]]
        prod = np.product(ns)
        print(prod, ns)
        if prod > maximum:
            maximum = prod

print("VERTICAL")
# vertical line loop
for col in nums.T:
    for i in range(len(col) - 3):
        ns = [col[i], col[i+1], col[i+2], col[i+3]]
        prod = np.product(ns)
        print(prod, ns)
        if prod > maximum:
            maximum = prod
            
print("DIAGONAL 1")
# first diagonal line loop
for i in range(len(nums) - 3):
    for j in range(len(nums[i]) - 3):
        ns = [nums[i, j], nums[i+1, j+1], nums[i+2, j+2], nums[i+3, j+3]]
        prod = np.product(ns)
        print(prod, ns)
        if prod > maximum:
            maximum = prod

print("DIAGONAL 2")
# second diagonal line loop
for i in range(len(nums) - 3):
    for j in range(3, len(nums[i])):
        ns = [nums[i, j], nums[i+1, j-1], nums[i+2, j-2], nums[i+3, j-3]]
        prod = np.product(ns)
        print(prod, ns)
        if prod > maximum:
            maximum = prod
print(maximum)
```
