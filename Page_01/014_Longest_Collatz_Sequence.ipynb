```python
def collatz_count(n):
    if n == 1:
        return 1
    
    if n % 2 == 0:
        return 1 + collatz_count(n / 2)
    else:
        return 1 + collatz_count(3 * n + 1)

longest = 0
best_start_num = None 
for i in range(1, 1_000_000):
    if i % 10000 == 0:
        print(f"step: {i}\t\tbest start num: {best_start_num}\t\tlongest:{longest}")
    n = i
    count = 1
    while n != 1:
        count += 1
        if n % 2 == 0:
            n /= 2
        else:
            n = 3 * n + 1
    
    if count > longest:
        longest = count
        best_start_num = i

best_start_num, longest    
```
