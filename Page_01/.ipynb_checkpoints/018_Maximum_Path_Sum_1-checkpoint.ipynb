```python
class Node():
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right
    def __str__(self):
        return f"Node:\n\tvalue:\t{self.value}\n\tleft:\t{'' if self.left is None else self.left.value}\n\tright:\t{'' if self.right is None else self.right.value}"
```

```python
nodes = [[Node(75)]]
for i in range(1, len(data) - 1):
    next_nodes = []
    for j in range(len(data[i])):
        new_node = Node(data[i][j])
        if j < len(nodes[i-1]):
            nodes[i-1][j].left = new_node
        if j > 0:
            nodes[i-1][j - 1].right = new_node
        next_nodes.append(new_node)
    nodes.append(next_nodes)
```

```python
root = nodes[0][0]
def multiplier(root, path=[], total=0):
    new_path = path + [root.value]
    new_total = total + root.value
    if root.left is None or root.right is None:
        return root.value
    
    return root.value + max(multiplier(root.left, new_path, new_total), multiplier(root.right, new_path, new_total))

multiplier(root) + 93
```
